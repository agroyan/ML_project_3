Удаляем столбец "Unnamed: 0", так как он не несет никакого смылса 

Обрабатываем столбец "class": "suicide" -> 1 и "non-suicide" -> 0

```python
# Удаляем НЕ буквы
def delete_not_alphabet(data, column):
    list_of_text = data[column]
    new_list_of_text = []
    for i in list_of_text:
        new_i = re.sub("[^a-zA-Z]", " ", i)
        new_list_of_text.append(new_i)
    data[column] = new_list_of_text
    return data
    
# Переводим слова в нижний регистр     
def lower(data, column):
    list_of_text = data[column]
    new_list_of_text = []
    for i in list_of_text:
        new_i = i.lower()
        new_list_of_text.append(new_i)
    data[column] = new_list_of_text
    return data

# Удаляем стоп слова и обрезаем остальные слова с помощью PorterStemmer   
def delete_stop_words_and_stemmer_words(data, column):
    porter = PorterStemmer()
    unique_stops = set(stopwords.words('english'))
    list_of_text = data[column]
    new_list_of_text = []
    for i in list_of_text:
        list_of_i = word_tokenize(i)
        new_list_of_i = []
        for element in list_of_i:
            if (element not in unique_stops):
                element = porter.stem(element)
                new_list_of_i.append(element)
        new_text = " ".join(new_list_of_i)
        new_list_of_text.append(new_text)
    data[column] = new_list_of_text
    return data
```

```python
data = delete_not_alphabet(data, "text")
data = lower(data, "text")
data = delete_stop_words_and_stemmer_words(data, "text")
```

Теперь векторизуем слова с помощью техники Word2Vec. Это техника воплощает идею контекстной близости. Потом обучим модели на этом векторном представлении.

```python
from gensim.models import Word2Vec
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

#токенизация 
sentences = list(data['text'])
L = []
for sentence in sentences:
    L.append(word_tokenize(sentence))
```

```python
#размер векторного представления 
size_n = 150
```

```python
#Word2Vec
model = Word2Vec(L, vector_size=size_n, window=5)
```

Теперь каждое предложение состоит из набора векторов. Как с этим работать? Ведь по сути нам каждое предложение нужно представить в виде одного вектора для работы со стандартными моделями. Решим эту проблему просто. Усредним векторы слов, которые имеют векторное представление. 

```python
def sen_to_vec(sentence):
    list_of_words = word_tokenize(sentence)
    result_vec = np.zeros(size_n, dtype=float)
    n = len(list_of_words)
    i = 0
    count = 0
    while (i < n):
        try:
            result_vec = result_vec + model.wv[list_of_words[i]]
            count += 1
        except KeyError:
            pass
        i += 1
    if (count == 0):
        return result_vec
    return result_vec / count
```

```python
new_data_np = np.zeros((len(sentences), size_n), dtype=float)
i = 0
for sentence in sentences:
    new_data_np[i] = sen_to_vec(sentence)
    i += 1
```

```python
new_data = pd.DataFrame(new_data_np)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_data, data["class"], test_size=0.3, random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(penalty=None, max_iter=130)
clf.fit(X_train, np.ravel(y_train))
y_predict = clf.predict(X_test)
y_train_new = clf.predict(X_train)
print("ACCURACY:", accuracy_score(y_predict, y_test))
print("RECALL", recall_score(y_test, y_predict))
```

```python
from catboost import CatBoostClassifier

clf = CatBoostClassifier(iterations=2000)
clf.fit(X_train, np.ravel(y_train))
y_predict = clf.predict(X_test)
y_train_new = clf.predict(X_train)
print("ACCURACY:", accuracy_score(y_predict, y_test))
print("RECALL", recall_score(y_test, y_predict))
```
